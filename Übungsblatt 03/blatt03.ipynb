```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
xy = np.array([[-2, 2], [-4, -2], [8, 2], [2, 4], [-2, 2]])

x, y = 6.9, 0.42

fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(xy[:, 0], xy[:, 1])
plt.scatter(x, y, marker="x", color="red", s=100, linewidths=4)
plt.xlabel("x")
plt.ylabel("y")

plt.show()

threshold = lambda x: int(x >= 0)

a1 = threshold(2*x - y + 6)
a2 = threshold(-(1/3*x - y - 2/3))
print(-(1/3*x - y - 2/3))
a3 = threshold(-1/3*x - y + 14/3)
a4 = threshold(1/2*x - y + 3)

output = threshold(a1+a2+a3+a4 - 4)

print("Bedingungen:", a1, a2, a3, a4)
print("Im Trapez:", output)
```
